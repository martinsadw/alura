# 1. Importing

```python
import pandas as pd
import sqlalchemy
```

```python
engine = sqlalchemy.create_engine('sqlite:///:memory:')
```

```python
clients = pd.read_csv('clientes_banco.csv', index_col='ID_Cliente')
clients
```

```python
clients.to_sql('clients', engine)
```

```python
inspector = sqlalchemy.inspect(engine)
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clients WHERE Categoria_de_renda = "Empregado"'
employee = pd.read_sql(query, engine, index_col='ID_Cliente')
employee
```

# 2. Exporting

```python
employee.to_sql('employee', engine)
```

```python
inspector = sqlalchemy.inspect(engine)
print(inspector.get_table_names())
```

```python
pd.read_sql_table('employee', engine, index_col='ID_Cliente')
```

```python
pd.read_sql_table('employee', engine, index_col='ID_Cliente', columns=['Grau_escolaridade', 'Rendimento_anual'])
```
