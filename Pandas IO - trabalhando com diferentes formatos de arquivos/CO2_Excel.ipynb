```python
# The data is heavily incorrect. It appears that zeros to the right were removed
# This code attempt to correct this by guesswork
countries = ['Brazil', 'United States']
a = emissoes.query('Country in @countries').drop(emissoes.columns[1:4], axis=1).set_index('Country').transpose()
a += a * 9 * (a < 100000000)
a += a * 9 * (a < 1000000000)
a['United States'] *= 10
a.plot()
```
