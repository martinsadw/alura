```python
import pandas as pd
```

# 1. Importing Emission

```python
excel_emission = pd.ExcelFile('emission.xlsx')
excel_emission.sheet_names
```

```python
pd.read_excel(excel_emission, 'GEE Estados', nrows=10)
```

# 2. Importing Population

```python
excel_population = pd.ExcelFile('population.xlsx')
excel_population.sheet_names
```

```python
data_population = pd.read_excel(excel_population, 'Municípios', skiprows=1, skipfooter=34, dtype={'POPULAÇÃO': str})
data_population['POPULAÇÃO'] = pd.to_numeric(data_population['POPULAÇÃO'].str.replace(r'\(\d+\)', '', regex=True).str.replace('.', ''))
data_population
```

```python
data_population.info()
```

# 3. Exporting

```python
data_emission = pd.read_excel(excel_emission, 'GEE Estados')
data_emission.to_csv('data_emission.csv', index=False)
```
