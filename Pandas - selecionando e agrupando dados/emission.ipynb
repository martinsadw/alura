# 2. Checking emission data

## 2.1. Checking "Emissão / Remoção / Bunker" consistency

```python
emission['Emissão / Remoção / Bunker'].unique()
```

Values of type "Remoção" correspond to removal and should be negative  
Values of type "Bunker" correspond to emission that doesn't have a state  
Values with "NCI" are data not included in the national inventory

### 2.1.1. Check if removal values are correct

Check if values marked as 'Remoção' are negative

```python
type_filter = ['Remoção', 'Remoção NCI']
emission.loc[emission['Emissão / Remoção / Bunker'].isin(type_filter), '1970':'2021'].max(axis=None)
```

### 2.1.2. Check if bunker values doesn't have a state

Check if values marked as "Bunker" don't have a state

```python
type_filter = ['Bunker']
emission.loc[emission['Emissão / Remoção / Bunker'].isin(type_filter), 'Estado'].unique()
```

# 3. Extracting relevant emission data

For this problem we're only intersted with the emission data included in the national inventory.  
This means that we're only keeping values of type "Emission"

```python
extracted_emission = emission.loc[emission['Emissão / Remoção / Bunker'] == 'Emissão']
extracted_emission = extracted_emission.drop('Emissão / Remoção / Bunker', axis=1)
extracted_emission
```

# 4. Formatting

```python
columns_info = extracted_emission.loc[:, 'Nível 1 - Setor':'Produto'].columns
columns_info
```

```python
columns_emission = extracted_emission.loc[:, '1970':'2021'].columns
columns_emission
```

```python
melted_emission = extracted_emission.melt(columns_info, columns_emission, 'Ano', 'Emissão')
melted_emission
```

```python
emission_by_gas = melted_emission.groupby('Gás')['Emissão'].sum().sort_values(ascending=False)
emission_by_gas.plot(kind='barh');
```

```python
co2_emission = emission_by_gas[:9].sum()
total_emission = emission_by_gas.sum()
print(f'Porcentagem de emissão de CO2 em relação ao total: {co2_emission / total_emission * 100:.2f}%')
```

# 5. Multiple index groupby

## 5.1. Cross-section

```python
sector_gas_emission = melted_emission.groupby(['Gás', 'Nível 1 - Setor'])[['Emissão']].sum()
sector_gas_emission
```

```python
sector_gas_emission.xs('CO2 (t)')
```

```python
sector_gas_emission.loc[('CO2 (t)', 'Mudança de Uso da Terra e Floresta')]
```

## 5.2. Max Emission by gas

```python
agg_columns = {
    "Emissão": pd.NamedAgg('Emissão', aggfunc='idxmax'),
    "Quantidade de emissão": pd.NamedAgg('Emissão', aggfunc='max'),
}
sector_gas_emission.groupby('Gás').agg(**agg_columns)
```

## 5.3. Max emission by sector

```python
agg_columns = {
    "Emissão": pd.NamedAgg('Emissão', aggfunc='idxmax'),
    "Quantidade de emissão": pd.NamedAgg('Emissão', aggfunc='max'),
}
sector_gas_emission.groupby('Nível 1 - Setor').agg(**agg_columns)
```

# 6. Population data

```python
population_by_state = population.groupby('UF')[['POPULAÇÃO']].sum()
population_by_state
```

# 7. Joining the data

```python
emission_by_state = melted_emission.query('Ano == "2021"').groupby('Estado')[['Emissão']].sum()
emission_by_state
```

```python
state_join = emission_by_state.join(population_by_state)
state_join
```

```python
state_join.plot(kind='scatter', x='POPULAÇÃO', y='Emissão');
```

```python
fig = px.scatter(state_join.reset_index(), x='POPULAÇÃO', y='Emissão', text='Estado')
fig.update_traces(textposition='top center')
```

```python
state_join['Emissão per capita'] = state_join['Emissão'] / state_join['POPULAÇÃO']
state_join.sort_values('Emissão per capita', ascending=False).plot(kind='bar', y='Emissão per capita', rot=0, width=0.8, figsize=(10, 4));
```

```python
fig = px.scatter(state_join.reset_index(), x='POPULAÇÃO', y='Emissão', text='Estado', size='Emissão per capita')
fig.update_traces(textposition='top center')
```
