```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# 1. Importing

```python
rent = pd.read_csv('aluguel.csv', sep=';')
rent
```

# 2. New Columns

```python
rent['Valor_por_mes'] = rent['Valor'] + rent['Condominio']
rent['Valor_por_ano'] = rent['Valor_por_mes'] * 12 + rent['IPTU']
rent
```

```python
def description(line):
    has_bedrooms = line["Quartos"] > 0
    has_garage = line["Vagas"] > 0
    text = f'{line["Tipo"]} em {line["Bairro"]}'
    if has_bedrooms or has_garage:
        text += ' com '
    if has_bedrooms:
        text += f'{line["Quartos"]} quarto{"" if line["Quartos"] == 1 else "s"}'
    if has_bedrooms and has_garage:
        text += ' e '
    if has_garage:
        text += f'{line["Vagas"]} vaga{"" if line["Vagas"] == 1 else "s"} de garagem'

    return text
rent['Descricao'] = rent.apply(description, axis=1)
rent
```

```python
rent['Possui_suite'] = rent['Suites'].apply(lambda x: 'Sim' if x > 0 else 'Não')
rent
```

# 3. Saving

```python
rent.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
