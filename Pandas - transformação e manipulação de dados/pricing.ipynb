## 0.1. dados_hospedagem.json

- `avaliacao_geral:` refere-se à média de notas dadas para a avaliação da hospedagem no imóvel.
- `experiencia_local:` descreve as experiências oferecidas durante a hospedagem no imóvel.

- `quantidade_quartos:` informa a quantidade de quartos disponíveis.
- `quantidade_camas:` informa a quantidade de camas disponíveis.
- `modelo_cama:` informa o modelo de cama oferecido.
- `comodidades:` informa as comodidades oferecidas pelo imóvel.
- `taxa_deposito:` informa a taxa de depósito mínima para segurança de hospedagem.
- `taxa_limpeza:` informa a taxa cobrada para o serviço de limpeza.
- `preco:` refere-se ao preço base a ser cobrado pela diária no imóvel.

# 1. Importing

## 1.1. dados_hospedagem.json

### 1.1.1. Format data

```python
import json
from pprint import pprint
with open('dados_hospedagem.json', 'r') as file:
    host_json = json.load(file)
```

```python
pprint(host_json, depth=1)
pprint(host_json['info_moveis'][0], depth=1)
```

```python
host = pd.json_normalize(host_json['info_moveis'])
host.head(3)
```

```python
# Explode lists into new rows
explode_columns = ['descricao_local', 'descricao_vizinhanca', 'quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas', 'modelo_cama', 'comodidades', 'taxa_deposito', 'taxa_limpeza', 'preco']
exploded_host = host.explode(explode_columns).reset_index(drop=True)
exploded_host.head(5)
```

```python
converted_host = exploded_host.copy()
```

### 1.1.2. Numeric data

```python
int_columns = ['max_hospedes', 'quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
float_columns = ['avaliacao_geral']
converted_host[int_columns] = exploded_host[int_columns].astype(int)
converted_host[float_columns] = exploded_host[float_columns].astype(float)
converted_host.info()
```

```python
price_columns = ['taxa_deposito', 'taxa_limpeza', 'preco']
converted_host[price_columns] = exploded_host[price_columns].applymap(lambda x: x.replace('$', '').replace(',', '').strip()).astype(float)
converted_host.info()
```

```python
converted_host.head(5)
```

### 1.1.3. Textual data

```python
converted_host['descricao_local'] = exploded_host['descricao_local']
converted_host['descricao_local'] = converted_host['descricao_local'].str.lower()
converted_host['descricao_local'] = converted_host['descricao_local'].str.replace('[^a-zA-Z0-9\-\' ]', '', regex=True)
converted_host['descricao_local'] = converted_host['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
converted_host['descricao_local'] = converted_host['descricao_local'].str.split()
converted_host['descricao_local']
```

```python
converted_host['comodidades'] = exploded_host['comodidades']
converted_host['comodidades'] = converted_host['comodidades'].str.replace('[{}"]', '', regex=True).str.split(',')
converted_host['comodidades']
```

```python
converted_host['descricao_vizinhanca'] = exploded_host['descricao_vizinhanca']
converted_host['descricao_vizinhanca'] = converted_host['descricao_vizinhanca'].str.lower()
converted_host['descricao_vizinhanca'] = converted_host['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\' ]', '', regex=True)
converted_host['descricao_vizinhanca'] = converted_host['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
converted_host['descricao_vizinhanca'] = converted_host['descricao_vizinhanca'].str.split()
converted_host['descricao_vizinhanca']
```
